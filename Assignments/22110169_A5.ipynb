```python
import math
import matplotlib.pyplot as plt
import numpy as np
eps=0.0002
Re=float(input("Enter Reynold's Number Re : "))

# defining the given function 
def g(x,Re):
    return 4*math.log10(Re*(x**0.5))-0.4-1/(x**0.5)

# plotting the function 
g_values=[]
d_values=[]
filtered_range = [x for x in range(0, 100) if x != 0]
for i in filtered_range:
    j=i/(10000)
    g_val=g(j,Re)
    g_values.append(g_val)
    d_values.append(j)
l=len(d_values)
plt.plot(d_values,g_values)
l=len(d_values)
plt.plot(d_values,np.zeros(l))
plt.xlabel("f")
plt.ylabel("g(f)")

# defining the derivative
def g_deri(x):
    return ((2/math.log(10)))/x+0.5/(x**1.5)

# Defining the Newton Raphson Method
def nr(zero,Re):
    zero= zero-g(zero,Re)/g_deri(zero)
    # print(zero)
    if abs(g(zero,Re))<eps:
        return zero
    else : 
        return nr(zero,Re)

# Calling the nr function to find the root
zero = nr(0.0075,Re)
print("Re : ", Re)
print("f for given Re is : ",zero)
# gor i in range(10000,500000):
#     print(i)
#     nr(0.0075,i)
```
