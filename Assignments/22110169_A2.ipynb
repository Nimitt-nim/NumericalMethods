```python
# Instructions : The following code calcultes all the zeros of f(v) and plots them with corresponding ranges

import matplotlib.pyplot as plt
import numpy as np
eps = 0.000001
a = 12.02 
b = 0.08407
R = 0.0821
T = 400
p = 2.5

def f(v):
    return (p + a/(v**2))*(v - b) - R*T

f_values=[]
d_values=[]
factor=100

filtered_range = [x for x in range(0, 130) if x != 0]
for i in filtered_range:
    j=i/10
    if f(j)<-50:
        f_val=-50
    else :
        f_val=f(j)
    f_values.append(f_val)
    d_values.append(j)
l=len(d_values)
plt.plot(d_values,f_values)
plt.plot(d_values,np.zeros(l),color='orange')
plt.title('f for v in range(0,13)')
plt.xlabel("v")
plt.ylabel("f (v)")
plt.show()







f_values=[]
d_values=[]
factor=100

filtered_range = [x for x in range(0, 100) if x != 0]
for i in filtered_range:
    j=i/300
    if f(j)<-50:
        f_val=-50
    else :
        f_val=f(j)
    f_values.append(f_val)
    d_values.append(j)
l=len(d_values)
plt.plot(d_values,f_values)
plt.plot(d_values,np.zeros(l),color='orange')
plt.plot((0.1,0.1),(10,-50),color='orange')
plt.plot((0.15,0.15),(10,-50),color='orange')
plt.plot((0.30,0.30),(10,-50),color='orange')
plt.title('f for v in range(0,0.3)')
plt.show()


def false_position(start,end,eps):
    l=f(start)
    u=f(end)
    zero=start-(l*(end-start))/(u-l)
    if abs(f(zero))<eps:
        return zero
    if f(zero)*f(start)<0:
        return false_position(start,zero,eps)
    else :
        return false_position(zero,end,eps)


start=0.1
end=0.15
zero = false_position(start,end,eps)
print("First zero: ",zero) 
print(f(zero))





f_values=[]
d_values=[]
factor=100

filtered_range = [x for x in range(3000, 4500) if x != 0]
for i in filtered_range:
    j=i/300
    if f(j)<-100:
        f_val=-100
    else :
        f_val=f(j)
    f_values.append(f_val)
    d_values.append(j)
l=len(d_values)
plt.plot(d_values,f_values)
plt.plot(d_values,np.zeros(l))
plt.title('f for v in range(12,13)')
plt.xlabel("v")
plt.ylabel("f (v)")
plt.show()


def false_position(start,end,eps):
    l=f(start)
    u=f(end)
    zero=start-(l*(end-start))/(u-l)
    if abs(f(zero))<eps:
        return zero
    if f(zero)*f(start)<0:
        return false_position(start,zero,eps)
    else :
        return false_position(zero,end,eps)

start=12
end=14
eps = 0.0000001
zero = false_position(start,end,eps)
print("root is : ",zero) 
print("error is : ",f(zero))

print("Only second zero i.e. 12.84 is only relevant to us")
```
